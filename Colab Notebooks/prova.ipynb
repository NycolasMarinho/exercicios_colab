```python
graus= float(input("Digite a temperatura em graus"))

Fahrenheit= graus * 1.8 + 32
Kelvin= graus + 273.15

print("a temperatura", graus, "graus em Fahrenheit é", Fahrenheit)
print("a temperatura", graus, "graus em Kelvin é", Kelvin)
```

exercicio 3

```python
num = int(input("Digite um numero: "))

i=1

while i * (i + 1) * (i + 2) < num:
   i += 1
if i * (i + 1) * (i + 2) == num:
    print(num, "é um numero triangular")
else:
    print(num, "não é um numero triangular")
```
