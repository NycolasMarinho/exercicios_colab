Estrutura de repetição

```python
inicio_contagem = int(input("Digite em que numero iniciará a contagem: "))
fim_contagem = int(input("Digite em que numero finalizará a contagem: "))
valor_somado = int(input("Digite o valor a ser somado: "))

contador = inicio_contagem
while contador <= fim_contagem:
  print(contador)
  contador = contador + valor_somado
print("Fim do programa")
```

A soma dos valores

```python
numeros_somados = int(input("Quantos numeros serão somados? "))
contador = 1
soma = 0
maior = 0

while contador <= numeros_somados:
  valor_que_sera_somado = int(input("Digite o valor a ser somado: "))
  if valor_que_sera_somado > maior:
    maior = valor_que_sera_somado
  soma = soma + valor_que_sera_somado
  contador = contador + 1

print("A soma dos valores é: ", soma)
print("O maior valor é: ", maior)
```

```python
numero_inicial = int(input("Digite o numero inicial: ")) #de quanto
numero_final = int(input("Digite o numero final: ")) #até quanto

if numero_final > numero_inicial:
  while numero_inicial <= numero_final:
    print(numero_inicial)
    numero_inicial = numero_inicial + 1

elif numero_final < numero_inicial:
  while numero_inicial >= numero_final:
    print(numero_inicial)
    numero_inicial = numero_inicial - 1
```

```python
quantidade_alunos = int(input("Quantos alunos serão digitados? "))
contador = 1
maior = 0
notas = []
alunos = []

while contador <= quantidade_alunos:
  nome_do_aluno = input(f"Digite o nome do aluno {contador}: ")
  nota_do_aluno = float(input(f"Digite a nota do aluno {contador}: "))
  notas.append(nota_do_aluno)
  alunos.append(nome_do_aluno)
  contador = contador + 1
  if nota_do_aluno > maior:
    maior = nota_do_aluno

if len(alunos) == len(notas):
    for aluno, nota in zip(alunos, notas):
        print(f"O aluno {aluno} tirou a nota {nota}")

print("A maior nota é: ", maior)
```
